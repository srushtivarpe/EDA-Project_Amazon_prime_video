## 3. ***Data Wrangling***

### Data Wrangling Code

#### Handling missing values

**1. titles.csv (dataset)**

````python
#find which columns has missing values
titles.isnull().sum()
````

````python
# find data type of each column
titles.dtypes
````

````python
# Fill missing values in titles on the basis of the data type of the column
# if data type is object use mode or use basic logics or domain knowledge to handle missing values
# if data type is integer or float use mean or median (which will be determined on the basis of the skew of that column)
# if skew is in the range -0.5 to 0.5  use mean
# if skew is not in the range -0.5 to 0.5 use median
# -----------column : 'age_certification'------------
titles.fillna({'age_certification':'Not Rated'},inplace=True)
# -----------column : 'description'------------
titles.fillna({'description':'Not provided'}, inplace=True)
# -----------column : 'seasons'------------
titles.fillna({'seasons':titles['seasons'].median()},inplace=True)  # as skew is 14.64
# -----------column : 'imdb_id'------------
titles.fillna({'imdb_id':'Not provided'},inplace=True)
# -----------column : 'imdb_score'------------
titles.fillna({'imdb_score':titles['imdb_score'].mean()},inplace=True)  # as skew is -0.35
# -----------column : 'imdb_votes'------------
titles.fillna({'imdb_votes':titles['imdb_votes'].median()},inplace=True)  # as skew is 12.51
# -----------column : 'tmdb_popularity'------------
titles.fillna({'tmdb_popularity':titles['tmdb_popularity'].median()},inplace=True)  # as skew is -26.77
# -----------column : 'score'------------
titles.fillna({'tmdb_score':titles['tmdb_score'].mean()},inplace=True)  # as skew is -0.335
````

#### **Chart - 2 Bar chart of genres vs. count**

````python
# Chart - 2 visualization code
# Extract genres
genre_list = []
for genres in titles['genres'].dropna():
#dropna() ensures that any missing (NaN) values are ignored.
    if genres.startswith('['):
        genre_list.extend(eval(genres))
# eval(genres) converts the string representation of a list (e.g., "[\"Drama\", \"Comedy\"]") into an actual Python list (["Drama", "Comedy"]).
#.extend() is used to add each genre from the list to genre_list.
    else:
# If the value does not start with [ (i.e., it is a single genre as a string, like "Drama"), it is added directly to genre_list using .append().
        genre_list.append(genres)

# Count genres
genre_counts = pd.Series(genre_list).value_counts()

# Plot chart
plt.figure(figsize=(12, 6))
sns.barplot(x=genre_counts.index, y=genre_counts.values, palette='viridis')
plt.xticks(rotation=45, ha='right')
plt.xlabel("Genres")
plt.ylabel("Count")
plt.title("Genre Distribution")
plt.show()
````

##### 1. Why did you pick the specific chart?


* To find the most common genres in the dataset.
* Bar chart is the most effective way to visualize categorical data (genres) against their frequency (count).

##### 2. What is/are the insight(s) found from the chart?

#### **Chart - 3 Boxplot of imdb_score across diferent genres**

````python
# Chart - 3 visualization code
import ast

# Expand the 'genres' column into individual rows
#Some movies/shows belong to multiple genres (e.g., ['Action', 'Thriller']).
#This step expands each genre into a separate row for proper analysis.
titles= titles.explode('genres')

# Convert the genre column from string representation of list to actual list
#The genres column contains genre data in string format (e.g., "[Action, Thriller]").
#ast.literal_eval() converts these strings into actual Python lists.
titles['genres'] = titles['genres'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) and x.startswith('[') else x)

# Further explode to ensure each genre has its own row
titles = titles.explode('genres')

# Drop rows with missing IMDb scores or genres
filtered = titles.dropna(subset=['imdb_score', 'genres'])

# Plot boxplot of IMDb score distribution across different genres
plt.figure(figsize=(12, 6))
sns.boxplot(data=filtered, x='genres', y='imdb_score',color='orange')
plt.xticks(rotation=90)
plt.xlabel("Genres")
plt.ylabel("IMDb Score")
plt.title("IMDb Score Distribution Across Different Genres")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()


````

#### **Chart - 6 Bar chart of genres vs. average runtime**

````python
# Chart - 6 visualization code
# Expand the 'genres' column into individual rows
#Some movies/shows belong to multiple genres (e.g., ['Action', 'Thriller']).
#This step expands each genre into a separate row for proper analysis.
titles= titles.explode('genres')

# Convert the genre column from string representation of list to actual list
#The genres column contains genre data in string format (e.g., "[Action, Thriller]").
#ast.literal_eval() converts these strings into actual Python lists.
titles['genres'] = titles['genres'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) and x.startswith('[') else x)

# Further explode to ensure each genre has its own row
titles = titles.explode('genres')

# Group by genre and calculate the average runtime
genre_runtime = titles.groupby("genres")["runtime"].mean()

# Plot bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x=genre_runtime.index, y=genre_runtime.values, palette="plasma")

# Labels and title
plt.xlabel("Genre")
plt.ylabel("Average Runtime (minutes)")
plt.title("Average Runtime of Movies and Shows by Genre")
plt.xticks(rotation=45, ha="right")

# Show plot
plt.show()
````
